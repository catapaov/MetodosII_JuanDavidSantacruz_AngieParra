```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solucion_analitica_1(x):
    return np.e**x
```

```python
def solucion_analitica_general(x,q):
    return ((1-q)*x+1)**(1/(1-q))
```

```python
def f(x,y,q):
    return y**q
```

```python
def GetEuler2(x,y,h,q):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1],q)
        k2 = f(x[i-1]+h,y[i-1] + h*k1,q)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
x = np.linspace(0,100,1000)
h = x[1] - x[0]

ya_0 = solucion_analitica_general(x,0)
ya_02 = solucion_analitica_general(x,0.2)
ya_04 = solucion_analitica_general(x,0.4)
ya_07 = solucion_analitica_general(x,0.7)
ya_09 = solucion_analitica_general(x,0.9)
ya_1 = solucion_analitica_1(x)

yn_0 = np.ones_like(x)
yn_02 = np.ones_like(x)
yn_04 = np.ones_like(x)
yn_07 = np.ones_like(x)
yn_09 = np.ones_like(x)
yn_1 = np.ones_like(x)
GetEuler2(x,yn_0,h,0)
GetEuler2(x,yn_02,h,0.2)
GetEuler2(x,yn_04,h,0.4)
GetEuler2(x,yn_07,h,0.7)
GetEuler2(x,yn_09,h,0.9)
GetEuler2(x,yn_1,h,1)
```

```python
plt.plot(x,ya_0,label="Solución analítica")
plt.plot(x,yn_0,label="Solución numérica")
plt.ylabel("u")
plt.xlabel("t")
plt.title("Comparación de la solución analítica contra la númerica para q = 0")
plt.legend()
plt.show()
```
