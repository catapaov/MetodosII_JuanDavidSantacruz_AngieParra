```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fax(x):
    return -x*np.pi**2
```

```python
def analytical(t):
    return (1/np.pi)*np.sin(np.pi*t)
```

```python
def GetVerlet(t,x,vx,h):
    v = vx + fax(x[0])*h    
    x[1] = x[0] + v*h
    
    for i in range(2,len(x)):
        a = fax(x[i-1])
        x[i] = 2*x[i-1]-x[i-2]+h**2*a
        
    
        
        
```

```python
t = np.linspace(0,10,100)
h = t[1]-t[0]
x = np.zeros_like(t)
vx = 1
GetVerlet(t,x,vx,h)
xa = analytical(t)
```

```python
plt.plot(t,x,label="Solucin nmerica")
plt.plot(t,xa,label="Solucin analtica")
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.title("Algoritmo de verlet para un oscilador armnico simple")
plt.legend()
```
