```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solucion_analitica(x):
    return (1/x**2)-2/(x**2+2)
```

```python
x = np.linspace(np.sqrt(2),100,300)
h = x[1]-x[0]
y = solucion_analitica(x)
```

```python
def f(x,y):
    return x*y**2-2*y/x -1/x**3
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
y_2 = np.zeros_like(x)
GetEuler2(x,y_2,h)
```

```python
plt.xlabel("t(s)")
plt.ylabel("y")
plt.title("Solución análitica y númerica a la ecuación diferencial de Riccati")
plt.plot(x,y,label="solucion_analitica")
plt.plot(x,y_2,label="Solución númerica")
plt.legend()
plt.show()
```
