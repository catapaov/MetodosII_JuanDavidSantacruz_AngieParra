```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def System(y,t,r=0.1,m=0.1,d=0.3,g=9.81):

    phi,psi,theta,dtheta = y


    Iz = (1/2)*m*r**2
    I0 = (1/4)*m*r**2+m*d**2
    dpsi = 400

    pphi = Iz*dpsi



    pphi = Iz*dpsi*np.cos(np.pi/4)

    dphi = (pphi-Iz*dpsi*np.cos(theta))/((I0*np.sin(theta)**2+Iz*np.cos(theta)**2))


    ddtheta = (dphi**2*np.sin(theta)*np.cos(theta)*(I0-Iz)-dphi*dpsi*Iz*np.sin(theta)+m*g*d*np.sin(theta))/I0

    return np.array([dphi,dpsi,dtheta,ddtheta])
```

```python
a = np.array([0.,0.,np.pi/4,0.])
t = np.linspace(0, 8, 1000)
solucion = odeint(System, a, t)
phi = solucion[:,0]
psi = solucion[:,1]
theta = solucion[:,2]
dtheta = solucion[:,3]
```

```python
r = 0.1
z = r*np.cos(theta)
y = r*np.sin(theta)*np.sin(phi)
x = r*np.sin(theta)*np.cos(phi)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Plot the trajectory in 3D
ax.scatter(x,y,z)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# Funcin para inicializar la animacin
def init():
    line.set_data([], [])
    line.set_3d_properties([])
    return line,

# Funcin para actualizar la animacin en cada cuadro
def update(frame):
    line.set_data(x[:frame], y[:frame])
    line.set_3d_properties(z[:frame])
    return line,

# Crear la figura y el objeto Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Configurar los lmites de los ejes
ax.set_xlim(min(x), max(x))
ax.set_ylim(min(y), max(y))
ax.set_zlim(min(z), max(z))

# Inicializar la lnea que ser animada
line, = ax.plot(x, y, z, lw=2)

# Crear la animacin
ani = FuncAnimation(fig, update, frames=len(x), init_func=init, blit=True)

# Mostrar la animacin
plt.show()
HTML(ani.to_html5_video())
```
