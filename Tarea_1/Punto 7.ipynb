```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def metodo_euler(u0,alfa,paso,iteraciones=100):
    y = np.zeros(iteraciones)
    x = np.zeros(iteraciones)
    up = u0
    for i in range (iteraciones):
        x[i] = 0 +(i)*paso
        y[i] = up*(1 + alfa*paso)**(i)
        up = y[i]
    return y,x
```

```python
y_1,x_1 = metodo_euler(1,-1,1.1)
y_2,x_1 = metodo_euler(1,-1,1.5)
y_3,x_1 = metodo_euler(1,-1,1.9)
plt.xlabel("segundos")
plt.ylabel("u")
plt.title("Soluciones númericas a la ecuación diferencial presentada")
plt.plot(x_1,y_1,label="paso de 1.1 segundos")
plt.plot(x_1,y_2,label="paso de 1.5 segundos")
plt.plot(x_1,y_3,label="paso de 1.9 segundos")
y_e = np.e**(-1*x_1)
plt.plot(x_1,y_e,label="solución exacta a la ecuación diferencial")
plt.legend()
plt.show()
```
