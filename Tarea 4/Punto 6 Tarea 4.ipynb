```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def numerov(r0,t,R):
    h = t[1]-t[0]
    psi = np.zeros_like(t)
    psi[0] = r0[0]
    psi[1] = r0[1]
    for i in range(2,len(t)):
        y2 = 1-(1/12)*h**2*R[i]
        y1 = 2*psi[i-1]*(1+((5/12)*h**2*R[i-1]))
        y0 = psi[i-2]*(1-(1/12)*h**2*R[i-2])
        psi[i] = (y1-y0)/y2
    return psi
        
```

```python
N = 1000
x = np.linspace(-5,5,N)
```

```python
def potencinho(E,x,m=1,w=1):
    return 2*((1/2)*m*w*x**2-E)
```

```python
r0 = (0,1*10**(-5))
E = np.linspace(0,6,601)

dE = 0.001
condicion_de_frontera = []
for i in range(len(E)):
    Vx = potencinho(E[i],x)
    solucinho_E = numerov(r0,x,Vx)
    VxDE = potencinho(E[i]+dE,x)

    solucinho_dE = numerov(r0,x,VxDE)
    if round(solucinho_E[-1]*solucinho_dE[-1],8) <0:
        b = round(E[i],3)
        print(f"La energía del estado ligado es:{b}")
        plt.plot(x,solucinho_E/np.max(solucinho_E),label=f"E={E[i]}")
plt.xlabel("x")
plt.ylabel(r"$\psi$(x)")
plt.legend()
```
